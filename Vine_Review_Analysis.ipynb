```python
# Add pandas dependency
import pandas as pd
```

```python
# Vine table file to load

vine_file= "vine_table.csv"
```

```python
# Read vine table

vine_df= pd.read_csv(vine_file)
vine_df.head()
```

```python
vine_df.dtypes
```

```python
total_votes_20_df= vine_df[vine_df['total_votes']>=20]
total_votes_20_df.head()
```

```python
vine_table_df= total_votes_20_df[(total_votes_20_df['helpful_votes']/total_votes_20_df['total_votes'])>=.50]
vine_table_df.head()
```

```python
vine_table_yes_df= vine_table_df[vine_table_df['vine'] == 'Y']
vine_table_yes_df.head()
```

```python
vine_table_no_df= vine_table_df[vine_table_df['vine'] == 'N']
vine_table_no_df.head()
```

```python
total_yes_count= vine_table_yes_df['star_rating'].count()
yes_star5_reviews_df=vine_table_yes_df[vine_table_yes_df['star_rating'] == 5]
yes_5star_count= yes_star5_reviews_df['star_rating'].count()
yes_percentage= yes_5star_count/total_yes_count * 100
```

```python
total_no_count= vine_table_no_df['star_rating'].count()
no_star5_reviews_df=vine_table_no_df[vine_table_no_df['star_rating'] == 5]
no_5star_count= no_star5_reviews_df['star_rating'].count()
no_percentage= no_5star_count/total_no_count * 100
```

```python
print(f'The total number of reviews that were made by people who paid the Vine program are {total_yes_count}.')
print(f'Of that total {yes_5star_count} gave a 5 star rating.')
print(f'The percentage of people who paid for Vine and gave a 5 star rating were {yes_percentage: .2f}%')
```

```python
print(f'The total number of reviews that were made by people who did not pay for the Vine program are {total_no_count}.')
print(f'Of that total {no_5star_count} gave a 5 star rating.')
print(f'The percentage of people who did not pay for Vine and gave a 5 star rating were {no_percentage: .2f}%')
```
